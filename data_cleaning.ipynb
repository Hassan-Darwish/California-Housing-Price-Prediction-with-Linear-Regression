```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X_mean = np.mean(df, axis=0)
X_std = np.std(df, axis=0)
df_normalized = (df-X_mean)/X_std
```

```python
df_normalized['ocean_proximity'] = df['ocean_proximity']
df_encoded = pd.get_dummies(df_normalized, columns=['ocean_proximity'])
df_encoded
```

```python
df_encoded.drop(["longitude", "latitude"], axis=1, inplace=True)

del df_normalized 
```

```python
df_encoded.columns
```

```python
df_encoded["median_income"] = df["median_income"]
df_encoded["income_cat"] = pd.cut(
    df_encoded["median_income"],
    bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
    labels=[1, 2, 3, 4, 5]
    )
df_encoded
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_idx, test_idx in split.split(df_encoded, df_encoded["income_cat"]):
    strat_train_set = df_encoded.loc[train_idx]
    strat_test_set = df_encoded.loc[test_idx]
    
```

```python
for dataset in (strat_train_set, strat_test_set):
    dataset.drop(['income_cat'], axis=1, inplace=True)
```

```python
X_mean = np.mean(df['median_income'])
X_std = np.std(df['median_income'])
strat_train_set['median_income'] = strat_test_set['median_income'] = (df['median_income']-X_mean)/X_std
```

```python
strat_train_set.to_csv('housing_training.csv')
strat_test_set.to_csv('housing_testing.csv')
```
